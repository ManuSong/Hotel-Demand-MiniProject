# 컬럼별 의미
- hotel - 호텔 종류 (H1 = 리조트 호텔 , H2= 시티 호텔)
- is_canceled- 예약이 취소 되었는지 아닌지를 0과 1로 구분
- lead_time - 소요시간 PMS에 예약이 입력된 날짜와 도착 날짜 사이에 경과된 일수
- arrival_date_year- 도착한 연도
- arrival_date_month- 도착한 달 
- arrival_date_week_number - 도착 날짜의 주 번호 
- arrival_date_day_of_month-- 도착일 
- stays_in_weekend_nights- 손님이 호텔에 숙박했거나 숙박을 예약한 주말 밤(토요일 또는 일요일)의 수  
- stays_in_week_nights- 게스트가 호텔에 숙박했거나 숙박을 예약한 주중 밤 수 (월요일~ 금요일)  
- adults 성인 수   
- children 어린이 수  
- babies 아기 수  
- meal 식사- 예약한 식사 종류, 예약한 식사 종류. 범주는 표준 환대 식사 패키지로 표시됩니다. 정의되지 않음/SC – 식사 패키지 없음; BB – 숙박 및 아침 식사; HB – 하프 보드(아침 식사 및 기타 식사 – 일반적으로 저녁 식사); FB – 풀보드(아침, 점심, 저녁)  
- country - 국가   
- market_segment 시장 - 시장 부문 지정 범주에서 "TA"라는 용어는 "Travel agency"를 의미하고 "TO"는 "Travel operator"를 의미합니다.  
- distribution_channel 유통 채널 - 예약 배포 채널. "TA"는 "Travel agency"를 의미하고 "TO"는 "Travel operator"를 의미합니다.  
- is_repeated_guest 예약 이름이 반복 손님의 것인지 (1) 아닌지 (0) 1과 0으로 표시  
- previous_cancellations  이전 취소 - 현재 예약 이전에 고객이 취소한 이전 예약 수    
- previous_bookings_not_canceled - 현재 예약 이전에 고객이 취소하지 않은 이전 예약 수  
- reserved_room_type - 예약된 객실 유형의 코드입니다. 익명성을 이유로 지정 대신 코드를 제시합니다.  
- assigned_room_type -예약에 할당된 객실 유형에 대한 코드입니다. 호텔 운영상의 이유(예: 오버부킹) 또는 고객의 요청으로 인해 배정된 객실 유형이 예약된 객실 유형과 다른 경우가 있습니다. 익명성을 이유로 지정 대신 코드를 제시합니다  
- booking_changes 예약 변경 - PMS에 예약이 입력된 시점부터 체크인 또는 취소 시점까지 예약에 대한 변경/수정 횟수
- deposit_type - 고객이 예약을 보장하기 위해 입금했는지 여부에 대한 표시. 이 변수는 세 가지 범주를 가정할 수 있습니다.  예치금 없음 – 예치금이 생성되지 않았습니다. 환불 불가 – 총 숙박 비용에 해당하는 보증금이 입금되었습니다. 환불 가능 – 총 숙박 비용에 미달하는 금액의 보증금이 입금되었습니다.  
- agent - 예약한 여행사의 ID  
- company - 예약을 했거나 예약 결제를 담당하는 회사/단체의 ID입니다. 익명성을 위해 지정 대신 ID를 제시함  
- days_in_waiting_list - 예약이 고객에게 확인되기 전에 대기자 명단에 있던 일수  
- customer_type -네 가지 범주 중 하나를 가정하는 예약 유형: 계약 - 예약에 관련된 할당 또는 기타 유형의 계약이 있는 경우 그룹 – 예약이 그룹과 연결된 경우 임시 – 예약이 그룹 또는 계약의 일부가 아니며 다른 임시 예약과 연결되지 않은 경우 일시적 당사자 – 예약이 일시적이지만 적어도 다른 임시 예약과 연결되어 있는 경우
- adr - 모든 숙박 거래의 합계를 총 숙박 일수로 나눈 평균 일 요금  
- required_car_parking_spaces - 고객이 필요로 하는 주차 공간 수  
- total_of_special_requests - 고객의 특별 요청 수(예: 트윈 침대 또는 고층)  
- reservation_status 예약 상태 - 세 가지 카테고리 중 하나를 가정한 예약 마지막 상태: 취소 – 고객이 예약을 취소했습니다. 체크아웃 – 고객이 체크인했지만 이미 출발했습니다. No-Show – 고객이 체크인하지 않았으며 그 이유를 호텔에 알렸습니다.  
- reservation_status_date 예약 상태 날짜 - 마지막 상태가 설정된 날짜입니다. 이 변수는 ReservationStatus와 함께 예약이 취소된 시점 또는 고객이 호텔에서 체크아웃한 시점을 파악하는 데 사용할 수 있습니다.

### 이상치 확인
- 어른이 0명인데 아이들 숫자가 있는경우? 있을 수 없다.
- 사람이 없는 경우??

```python
no_adult = (df['adults'] == 0) & (df['children'] >= 1) & (df['babies'] >= 1)
```

```python
no_person = (df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)
```

- 사람이 없는 경우가 있으니 제외
- 어른이 0명인데 투숙이 불가능한 경우가 있으니 이것도 이상치로 제외

```python
data = df[~(no_adult) | (no_person)]
```
